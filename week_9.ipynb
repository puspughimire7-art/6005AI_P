```python
# Install mlxtend (used for Apriori and association rules)
!pip install mlxtend --quiet

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Step 1: Define the transactions
transactions = [
    ['Mango', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],  # T1
    ['Doll', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],   # T2
    ['Mango', 'Apple', 'Key-chain', 'Eggs'],                       # T3
    ['Mango', 'Umbrella', 'Corn', 'Key-chain', 'Yo-yo'],           # T4
    ['Corn', 'Onion', 'Key-chain', 'Ice-cream', 'Eggs']            # T5
]
transactions
```

```python
# Step 2: Convert transactions to one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# Step 3: Run Apriori to find frequent itemsets
# Minimum support = 60% -> 0.6
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets.sort_values('support', ascending=False).reset_index(drop=True)
```

```python
# Step 4: Generate association rules
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.01)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
# Step 5: Filter strong rules (confidence >= 0.8)
strong_rules = rules[rules['confidence'] >= 0.8].copy()
strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```
